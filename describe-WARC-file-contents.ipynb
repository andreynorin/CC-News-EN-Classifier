```python
from numpy import int64

import pandas as pd
from warcio.archiveiterator import ArchiveIterator
from bs4 import BeautifulSoup

def build_titles_df():
    
    with open('CC-NEWS-20161119132828-00008.warc.gz', 'rb') as stream:

            recordCounter = 0

            for record in ArchiveIterator(stream):

                    if record.rec_type == 'response':
                        payload_content = record.raw_stream.read()
                        soup            = BeautifulSoup(payload_content, 'html.parser')
                        links           = soup.find_all('a')
                        linkCount       = len(links)
                        paragraphs      = soup.find_all('p')
                        paragraphCount  = len(paragraphs)
                        h1s             = soup.find_all('h1')
                        h1Count         = len(h1s)
                        h2s             = soup.find_all('h2')
                        h2Count         = len(h2s)
                        h3s             = soup.find_all('h3')
                        h3Count         = len(h3s)
                        h4s             = soup.find_all('h4')
                        h4Count         = len(h4s)
                        divs            = soup.find_all('div')
                        divCount        = len(divs)
                        url             = record.rec_headers.get_header('WARC-Target-URI')
                        contentLen      = record.rec_headers.get_header('Content-Length')
                        if (soup.title is not None):
                            title = soup.title.string
                            df.loc[recordCounter] = [title, url, contentLen, linkCount, paragraphCount, h1Count, h2Count, h3Count, h4Count, divCount]

                    recordCounter += 1
                    if recordCounter >= 10001:
                        break


    df.head()

df = pd.DataFrame(columns=(['Title','URL','ContentLength', 'linkCount', 'paragraphCount', 'h1Count', 'h2Count', 'h3Count', 'h4Count', 'divCount']))

build_titles_df()
```
