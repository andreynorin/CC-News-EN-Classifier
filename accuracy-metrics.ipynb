```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import warnings
import sklearn.exceptions
warnings.filterwarnings("ignore", category=sklearn.exceptions.UndefinedMetricWarning)
```

```python
# ignore the following warning "UndefinedMetricWarning: Precision and F-score are ill-defined 
# and being set to 0.0 in labels with no predicted samples. Use `zero_division` parameter to 
# control this behavior."

warnings.filterwarnings("ignore", category=sklearn.exceptions.UndefinedMetricWarning)

# build a classification report
preds_df = pd.read_parquet("accuracy_test_dataset_160k_training_examples.parquet")
preds_df['result'] = preds_df['result'].apply(lambda x : x[0])
print (classification_report(preds_df['topic'], preds_df['result']))
```

```python
# Plot confusion matrix
fig, ax = plt.subplots(figsize=(15, 9))
cmp = ConfusionMatrixDisplay(
    confusion_matrix(preds_df['topic'], preds_df['result']),
    display_labels=['BUSINESS','ENTERTAINMENT','HEALTH','NATION','SCIENCE','SPORT','TECHNOLOGY','WORLD'],
)

cmp.plot(ax=ax)
plt.xticks(rotation=45)
plt.show();
```
